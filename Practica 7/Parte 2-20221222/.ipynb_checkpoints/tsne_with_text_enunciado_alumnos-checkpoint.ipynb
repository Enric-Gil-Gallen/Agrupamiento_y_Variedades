
# t-SNE y la 'bag of words'
En esta práctica vamos a explorar el uso de t-SNE sobre la estructura de datos conocida como 'bag of words' utilizada para representar ficheros de texto. La estructura 'Bag of words' representa a cada fichero texto (articulo, libro, página de internet,..) como un vector de frecuencias. Cada componente del vector expresa la frecuencia de aparición de una palabra en el texto a que se refiere el vector. Los vectores se agrupan en matrices, de tal manera que cada fila corresponde a un texto.
Tenemos así, un conjunto de puntos en un espacio euclídeo altamente dimensional (el número de palabras puede ser todas las del diccionario) que puede intentar visualizarse con t-SNE para buscar estructura en el dataset (agrupamientos).

Disponemos de un dataset llamado 'wikipedia-vectors.csv' que contiene 13126 textos (filas) definidos por las frecuencias de 60 palabras (la primera columna es el número identificador del texto).

Vamos a cargarlo como un Pandas DataFrame utilizando su método 'read_csv' pero como el dataset es muy grande vamos a quedarnos con los 500 primeros textos. El dataset está barajado, así que entre los 500 hay una muestra homogénea de la población.

**ATENCIÓN: Elimina de este conjunto las filas que tengan todas las componentes a '0'** 

```python
import pandas as pd
pd.options.mode.chained_assignment = None 
import numpy as np
import re
import nltk
import sklearn.preprocessing as preproc


from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
%matplotlib inline

#lee aquí con pandas el archivo y usa los artibutos de lectura para establecer que lea solo los 500 primeros
data = pd.read_csv('./wikipedia-vectors.csv')
data = data.sample(n=data.shape[0])
data = data.head(500)

#Elimina las filas con todos los valores a '0', ya que son datos inútiles
data2 = data.drop(columns=['Unnamed: 0'])
data2 = data2[data2.sum(axis=1) != 0]
data = data[data.index.isin(data2.index)]
```

```python
#Guardo la primera columna de data en docs (que corresponde al identificador de los textos)
docs = data['Unnamed: 0']
#Elimino la primera columna de los datos (que corresponde al identificador de los textos)
data = data.drop(columns=['Unnamed: 0'])

#creo ahora docs2 , un DataFrame que contiene como primera columna docs y como segunda la palabra de más frecuencia de cada texto
#creación de docs2 con la primera columna
pal = []
for i in range(data.shape[0]):
    idx = (data.iloc[i]).idxmax()
    pal.append(idx)

datos = {'Id_doc':  docs,
        'Index_Max_Pals': pal
        }
docs2=pd.DataFrame(datos)

print(docs2)
```

**Actividad 1** Aquí vamos a bifurcar para experimentar. Primero ejecuta todo el resto de las boxes con el dataset llamado 'data' tal cual.
En una segunda ejecución vamos a normalizar los datos por filas (Usando Normalice). Esto reescala las características. Debes juzgar en tus conclusiones el efecto que esto ha causado sobre tus resultados.

```python
from sklearn.preprocessing import normalize
#Aquí en una primera ejecución usamos 'data'. En una segunda ejecución usaremos 'X' que obtenemos aquí
X = data.values
#normaliza los datos
X = normalize(X)
```

Vamos a hora a ejecutar t-SNE con sus parámetros. Fíjate que imprimimos su atributo kl_divergence_
Vamos a hacer un analisis de cómo varía este parámetro respecto de los valores de perplexity que como sabes es un hiperparámetro que suele ir entre 5 y 50.
Realiza un bucle de ejecucion de 45 t-SNEs adquiriendo la kl_divergence_ y luego ploteando la curva. Debes comentar la forma de la curva que obtienes. (Nota: esto **NO** es una técnica al estilo de la 'elbow curve' de k-means con el objetivo de seleccionar la perplexity. Solo tiene el objetivo de que ganes intuición del comportamiento de t-SNE al variar ese parámetro).

```python
#genera el modelo de t-SNE especificando por tu cuenta (no por defecto) los parámetros  perplexity, n_iter, early_exaggeration
#pon n_components=2 y random_state=23 para poderlo contrastar con mis resultados 

kl_div = []
for p in range(5,50+1):
    tsne_model = TSNE(perplexity = p, n_iter = 1000, early_exaggeration = 12, n_components=2, random_state=23)
    #aplica aquí el metodo fit_transform. Esto puede tardar un poco de tiempo
    new_values = tsne_model.fit_transform(data)

    kl_div.append(tsne_model.kl_divergence_)
```

```python
plt.plot(kl_div)
plt.xlabel('Perplexity')
plt.ylabel('kl_divergence_')
plt.show()
```

<font color='blue'>
    Se puede observar en la gráfica que, en las primeras ejecuciones, el valor de kl_divergence_ aumenta. Sin embargo, a partir de perplexity = 6 (aproximadamente), la divergencia va disminuyendo con ciertos altibajos.
</font>

Ahora vuelve a ejecutar  t-SNE con perplexity 50 y carga los resultados en x e y

```python
tsne_model = TSNE(perplexity = 50, n_iter = 1000, early_exaggeration = 12, n_components=2, random_state=23)
new_values = tsne_model.fit_transform(data)
x = []
y = []
for value in new_values:
    x.append(value[0])
    y.append(value[1])
```

Visualiza ahora el conjunto de puntos con algunas palabras identificadores de puntos para comprender como se agrupan los puntos. Puedes repetir el proceso tantas veces como consideres  modificando los hiperparámetros en los rangos que discutimos en clase.
